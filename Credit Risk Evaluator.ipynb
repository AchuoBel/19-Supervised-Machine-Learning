```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'), index_col= 0).drop(["index"], axis=1)
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'), index_col= 0).drop(["index"], axis=1)
```

```python
train_df
```

```python
# Convert categorical data to numeric and separate target feature for training data
training_y = train_df.loan_status
training_X = train_df.drop(columns = ["loan_status"])
training_X = pd.get_dummies(training_X)
training_X
```

```python
# Convert categorical data to numeric and separate target feature for testing data
testing_y = test_df.loan_status
testing_X = test_df.drop(columns = ["loan_status"])
testing_X = pd.get_dummies(testing_X)
testing_X
```

```python
# add missing dummy variables to testing set
cols_test = list(testing_X)
cols_train = list(training_X)
for column_name in cols_train:
    if column_name not in cols_test:
        # add that column to test_df
        testing_X[column_name] = 0
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
# Create a logistic regression model
lrg = LogisticRegression(max_iter=1000)

# Fit our model using the training data
lrg.fit(training_X, training_y)
lrg.score(testing_X, testing_y)
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
lrg = RandomForestClassifier()
lrg.fit(training_X, training_y)
lrg.score(testing_X, testing_y)
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(training_X)
training_X_scaled = scaler.transform(training_X)
testing_X_scaled = scaler.transform(testing_X)
```

```python
training_X_scaled
```

```python
testing_X_scaled
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
from sklearn.linear_model import LogisticRegression
lrg = LogisticRegression(solver='lbfgs')
lrg.fit(training_X_scaled, training_y)
lrg.score(testing_X_scaled,testing_y)
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
from sklearn.ensemble import RandomForestClassifier
lrg = RandomForestClassifier()
lrg.fit(training_X_scaled,  training_y)
lrg.score(testing_X_scaled, testing_y)
```
